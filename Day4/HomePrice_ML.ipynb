```python
import pandas as pd
```

```python
df= pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
df['area'],df['price']
```

```python
plt.scatter(df['area'],df['price'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(df[['area']],df['price'])
```

```python
print(lr.predict([[4800]]))
```

```python
print(lr.coef_, lr.intercept_)
```

```python
plt.scatter(df['area'], df['price'])
plt.plot(df['area'], lr.predict(df[['area']]), color='red')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("mean_squared_error:", mean_squared_error(df['price'], lr.predict(df[['area']])))
```

```python
print("R2_score:", r2_score(df['price'], lr.predict(df[['area']])))
```
